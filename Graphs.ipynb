```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file into a pandas DataFrame
df = pd.read_csv('cars_raw.csv')  # replace with your actual file path

# Step 2: Count the frequency of each car name
car_counts = df['Make'].value_counts()  # Replace 'CarName' with the actual name of the column

# Step 3: Plot the bar chart
car_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Step 4: Add labels and title
plt.title('Company Name')
plt.xlabel('Companies')
plt.ylabel('Number of Cars')
plt.xticks(rotation=90)  # Rotate x-axis labels if they are long

# Step 5: Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Read the CSV file into a pandas DataFrame
df = pd.read_csv('cars_raw.csv')  # Replace with your actual file path

# Step 2: Select the specific columns (Replace with actual column names)
selected_columns = df[['Mileage', 'Price', 'SellerRating']]  # Choose the columns you want

# Step 3: Compute the correlation matrix for the selected columns
correlation_matrix = selected_columns.corr()

# Step 4: Create a heatmap using Seaborn
plt.figure(figsize=(6, 5))  # Optional: Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Step 5: Add title and show the plot
plt.title('Correlation Heatmap among Mileage, Price and Seller Ratings')
plt.show()

```
